```python
import ipywidgets as widgets
from IPython.display import display
from joblib import load
import datetime  # Import the datetime module

# Load the saved model
loaded_model = load('/content/drive/MyDrive/linear_regression_model.joblib')

# Create input widgets for year, month, and day
year_widget = widgets.IntText(description='Year:', min=2000, max=2025, value=2024)
month_widget = widgets.IntText(description='Month:', min=1, max=12, value=1)
day_widget = widgets.IntText(description='Day:', min=1, max=31, value=1)

# Create a button widget for prediction
predict_button = widgets.Button(description='Predict')

# Create an output widget to display predictions
output_widget = widgets.Output()

# Define a function to make predictions
def predict_gold_price(user_year, user_month, user_day):
    # Convert date to days since a reference date (e.g., January 1, 1970)
    reference_date = datetime.datetime(1970, 1, 1)
    selected_date = datetime.datetime(user_year, user_month, user_day)
    days_since_reference = (selected_date - reference_date).days

    # Make prediction
    prediction = loaded_model.predict([[days_since_reference]])
    return prediction[0]

# Define a function to handle button click event
def predict_clicked(b):
    with output_widget:
        output_widget.clear_output()
        # Extract input values from widgets
        user_year = year_widget.value
        user_month = month_widget.value
        user_day = day_widget.value

        # Make prediction
        predicted_price = predict_gold_price(user_year, user_month, user_day)

        # Display the prediction
        print(f"Predicted gold price for {user_year}-{user_month}-{user_day}: ${predicted_price:.2f}")

# Attach the predict_clicked function to the button's click event
predict_button.on_click(predict_clicked)

# Display the widgets
display(year_widget, month_widget, day_widget, predict_button, output_widget)
```
